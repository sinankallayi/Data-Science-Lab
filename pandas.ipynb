```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print('\n',df) 
#refer to the row index:
print('\n',df.loc[0])
#use a list of indexes:
print('\n',df.loc[[0, 1]])
```

```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

print('\n',df)
#refer to the named index:
print('\n',df.loc["day2"])
```

```python
import pandas as pd

df = pd.read_csv('data.csv')

for x in df.index:
      if df.loc[x, "Duration"] > 120:
        df.loc[x, "Duration"] = 120

print(df.to_string())
```

```python
import pandas as pd

df = pd.read_csv('data.csv')

for x in df.index:
      if df.loc[x, "Duration"] > 120:
        df.drop(x, inplace = True)

print(df.to_string())
```
