```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Implement K-means clustering with explicit n_init parameter
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
kmeans.fit(X_scaled)

# Get the cluster labels
labels = kmeans.labels_

# Create a DataFrame for easier plotting
df = pd.DataFrame(X_scaled, columns=iris.feature_names)
df['Cluster'] = labels

# Plot the clusters
plt.figure(figsize=(10, 6))
plt.scatter(df[df['Cluster'] == 0]['sepal length (cm)'], df[df['Cluster'] == 0]['sepal width (cm)'], label='Cluster 1', s=100)
plt.scatter(df[df['Cluster'] == 1]['sepal length (cm)'], df[df['Cluster'] == 1]['sepal width (cm)'], label='Cluster 2', s=100)
plt.scatter(df[df['Cluster'] == 2]['sepal length (cm)'], df[df['Cluster'] == 2]['sepal width (cm)'], label='Cluster 3', s=100)

# Plotting centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=300, alpha=0.6, label='Centroids')

plt.title('K-means Clustering of Iris Dataset')
plt.xlabel('Sepal Length (standardized)')
plt.ylabel('Sepal Width (standardized)')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data

# Standardize the features
X_scaled = StandardScaler().fit_transform(X)

# K-means clustering
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
kmeans.fit(X_scaled)

# Plotting the clusters
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis', s=100, alpha=0.6)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=300, alpha=0.6, label='Centroids')

plt.title('K-means Clustering of Iris Dataset')
plt.xlabel('Sepal Length (standardized)')
plt.ylabel('Sepal Width (standardized)')
plt.legend()
plt.grid()
plt.show()
```
